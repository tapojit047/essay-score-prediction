```python
import pandas as pd

df = pd.read_excel("/Users/tapajit/projects/essay-score-prediction/dataset/essay_dataset_v1.xlsx")
```

```python
cols_meta = df.iloc[:, 0:5]
y = df.iloc[:, 1].values   # column B
```

```python
X_counts = df.iloc[:, 5:].values
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf = TfidfTransformer()
X_tfidf = tfidf.fit_transform(X_counts).toarray()
```

```python
import pandas as pd

df_tfidf = pd.concat(
    [cols_meta, pd.DataFrame(X_tfidf, index=df.index)],
    axis=1
)

df_tfidf.to_excel("dataset_after_tfidf.xlsx", index=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_tfidf)

pca = PCA(n_components=50)  # choose number of components you want
X_pca = pca.fit_transform(X_scaled)

df_pca = pd.concat(
    [cols_meta, pd.DataFrame(X_pca, index=df.index)],
    axis=1
)

# Save as Excel
df_pca.to_excel("dataset_after_pca.xlsx", index=False)

print("Saved → dataset_after_pca.xlsx")
```
